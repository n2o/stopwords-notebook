```python
with open('exported-discussion.txt') as f:
    text = f.read()
```

```python
import nltk
import nltk.corpus as corpus
nltk.download("stopwords")
```

```python
import numpy as np
x, y = np.ogrid[:1000, :1000]

mask = (x - 500) ** 2 + (y - 500) ** 2 > 400 ** 2
mask = 255 * mask.astype(int)
```

```python
from wordcloud import WordCloud, STOPWORDS

current_stopwords = set(STOPWORDS)
current_stopwords.update(corpus.stopwords.words("german"))
current_stopwords.update(["mal", "ja", "aber", "schon", "w√§re", "gibt"])

wordcloud = WordCloud(background_color="white", mask=mask, width=1000, height=1000, stopwords=current_stopwords).generate(text)
wordcloud.to_file("wordcloud.png")
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
